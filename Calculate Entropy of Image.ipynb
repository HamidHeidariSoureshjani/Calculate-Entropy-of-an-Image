```python
import numpy as np
import matplotlib.pyplot as plt
import skimage.measure
```

```python
image1 = plt.imread('/content/1.jpeg')
entropy1 = skimage.measure.shannon_entropy(image1)

image2 = plt.imread('/content/2.jpg')
entropy2 = skimage.measure.shannon_entropy(image2)

image3 = plt.imread('/content/3.jpeg')
entropy3 = skimage.measure.shannon_entropy(image3)

image4 = plt.imread('/content/4.jpeg')
entropy4 = skimage.measure.shannon_entropy(image4)

image5 = plt.imread('/content/Lena.png')
entropy5 = skimage.measure.shannon_entropy(image5)

image6 = plt.imread('/content/baboon.jpeg')
entropy6 = skimage.measure.shannon_entropy(image6)
```

```python
# Plot the results using subplots
fig, axs = plt.subplots(2, 3)

axs[0, 0].imshow(image1, cmap='gray')
axs[0, 0].set_title(f'Entropy: {entropy1:.2f}')

axs[0, 1].imshow(image2, cmap='gray')
axs[0, 1].set_title(f'Entropy: {entropy2:.2f}')

axs[0, 2].imshow(image5, cmap='gray')
axs[0, 2].set_title(f'Entropy: {entropy5:.2f}')

axs[1, 0].imshow(image3, cmap='gray')
axs[1, 0].set_title(f'Entropy: {entropy3:.2f}')

axs[1, 1].imshow(image4, cmap='gray')
axs[1, 1].set_title(f'Entropy: {entropy4:.2f}')

axs[1, 2].imshow(image6, cmap='gray')
axs[1, 2].set_title(f'Entropy: {entropy6:.2f}')

for ax in axs.flat:
    ax.label_outer()

plt.show()
```
